# The Grammar of Graphics and plotnine

Inspired by this [Medium Article](https://towardsdatascience.com/a-comprehensive-guide-to-the-grammar-of-graphics-for-effective-visualization-of-multi-dimensional-1f92b4ed4149)

```python
from plotnine.data import mtcars

mtcars.head()
```

```python
from plotnine import ggplot, aes, geom_point, theme_bw
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg')) 
        + geom_point()
        + theme_bw())
```

```python
(ggplot(mtcars, 
            aes('wt', 'mpg', color='factor(gear)'))
            + geom_point()
            + theme_bw())
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)', size='cyl'))
        + geom_point() 
        + theme_bw())
```

```python
from plotnine import facet_wrap
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)'))
        + geom_point() 
        + facet_wrap('~cyl') 
        + theme_bw())
```

```python
from plotnine import stat_smooth
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)'))
        + geom_point() 
        + stat_smooth(method='lm') 
        + theme_bw())
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)', 
            size='cyl'))
        + geom_point() 
        + facet_wrap('~am') 
        + theme_bw())
```

```python
from plotnine import facet_grid
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)', 
            size='cyl'))
        + geom_point() 
        + facet_grid('am ~ carb') 
        + theme_bw())
```
